```python
import pandas as pd
import numpy as np
```

```python
polygons = pd.read_csv("../data/linnamagede_ruudunumbrid.csv")
```

```python
polygons.rename(columns={"Ruudunumber": "Ruudunumber(1:2000)"}, inplace=True)
```
